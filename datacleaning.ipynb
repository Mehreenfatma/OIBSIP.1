```python
# 1. Remove duplicates
data = data.drop_duplicates()
```

```python
# 2. Check and handle missing values
# View missing values
missing_values = data.isnull().sum()
# Replace 'target_column' and 'some_column' with actual column names from your data
# Example: Drop rows with missing 'price' values, or fill missing values in 'reviews_per_month'
# Assuming 'price' is your target column and 'reviews_per_month' has missing values
data = data.dropna(subset=['price'])  # Replace 'price' with your actual target column name
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].median())  # Replace with your actual column name


```

```python
# 3. Convert data types if needed
# Example: Convert date column to datetime
# Assuming you have a 'last_review' column with dates
data['last_review'] = pd.to_datetime(data['last_review'])  # Replace 'last_review' with your actual date column name
```

```python
# 4. Handle outliers
# Example: Remove outliers in numeric data based on z-score
from scipy import stats
import numpy as np # Import numpy and assign it to the alias 'np'
# Assuming 'minimum_nights' is a numeric column with potential outliers
data = data[(np.abs(stats.zscore(data['minimum_nights'])) < 3)]  # Replace 'minimum_nights' with your actual numeric column name
```

```python
# 5. Standardize formats for categorical data
# Assuming you have a 'neighbourhood_group' column with categorical data
data['neighbourhood_group'] = data['neighbourhood_group'].str.lower().str.strip()  # Replace 'neighbourhood_group' with your actual categorical column name

# Display cleaned data
print(data.head())
```
