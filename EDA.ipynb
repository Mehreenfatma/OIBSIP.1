```python
 #Step 1: Descriptive statistics
desc_stats = data.describe()

# Step 2: Distribution plots for key columns: Calories, Total Fat, Sugars, and Protein
fig, axs = plt.subplots(2, 2)
sns.histplot(data['Calories'], kde=True, color='skyblue', ax=axs[0, 0]).set_title('Calories Distribution')
sns.histplot(data['Total Fat'], kde=True, color='salmon', ax=axs[0, 1]).set_title('Total Fat Distribution')
sns.histplot(data['Sugars'], kde=True, color='lightgreen', ax=axs[1, 0]).set_title('Sugars Distribution')
sns.histplot(data['Protein'], kde=True, color='lightcoral', ax=axs[1, 1]).set_title('Protein Distribution')
plt.tight_layout()

desc_stats
```

```python
# Step 3: Correlation matrix for key numerical columns
correlation_matrix = data[['Calories', 'Total Fat', 'Sugars', 'Protein', 'Sodium', 'Carbohydrates']].corr()

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix for Key Nutritional Values")
plt.show()

correlation_matrix
```

```python
# Step 4: Analyzing average nutritional values by food category
category_means = data.groupby('Category')[['Calories', 'Total Fat', 'Sugars', 'Protein', 'Sodium']].mean()

# Plotting average nutrition by category
category_means.plot(kind='bar', stacked=False, figsize=(14, 8), colormap='viridis')
plt.title("Average Nutritional Values by Food Category")
plt.ylabel("Average Amount")
plt.xticks(rotation=45)
plt.legend(title='Nutritional Values')
plt.show()

category_means
```

```python
# Calculating IQR for calories as an example
Q1 = data['Calories'].quantile(0.25)
Q3 = data['Calories'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['Calories'] > Q3 + 1.5 * IQR) | (data['Calories'] < Q1 - 1.5 * IQR)]

outliers[['Category', 'Item', 'Calories']]
```

```python
# Example for analyzing correlation in one category (e.g., 'Breakfast')
breakfast_data = data[data['Category'] == 'Breakfast']
breakfast_corr = breakfast_data[['Calories', 'Total Fat', 'Sugars', 'Protein', 'Sodium']].corr()

import seaborn as sns
sns.heatmap(breakfast_corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Breakfast Items")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot size and style
plt.figure(figsize=(14, 6))
sns.set(style="whitegrid")

# Scatter plot and regression for Calories vs Total Fat
plt.subplot(1, 2, 1)
sns.regplot(data=data, x='Total Fat', y='Calories', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Calories vs. Total Fat")
plt.xlabel("Total Fat (g)")
plt.ylabel("Calories")

# Scatter plot and regression for Calories vs Protein
plt.subplot(1, 2, 2)
sns.regplot(data=data, x='Protein', y='Calories', scatter_kws={'alpha':0.5}, line_kws={'color':'blue'})
plt.title("Calories vs. Protein")
plt.xlabel("Protein (g)")
plt.ylabel("Calories")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Calories vs Total Fat by Category
sns.lmplot(data=data, x='Total Fat', y='Calories', hue='Category', aspect=1.2, height=6, scatter_kws={'alpha':0.5})
plt.title("Calories vs. Total Fat by Category")
plt.xlabel("Total Fat (g)")
plt.ylabel("Calories")

# Calories vs Protein by Category
sns.lmplot(data=data, x='Protein', y='Calories', hue='Category', aspect=1.2, height=6, scatter_kws={'alpha':0.5})
plt.title("Calories vs. Protein by Category")
plt.xlabel("Protein (g)")
plt.ylabel("Calories")

plt.show()
```
