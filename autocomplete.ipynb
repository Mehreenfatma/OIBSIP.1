```python
import time
from collections import defaultdict, Counter
from sklearn.metrics import accuracy_score
```

```python
# N-gram-based Autocomplete
def build_ngrams(texts, n=2):
    ngrams = defaultdict(Counter)
    for text in texts:
        words = text.split()
        for i in range(len(words) - n + 1):
            ngram = tuple(words[i:i + n - 1])
            next_word = words[i + n - 1]
            ngrams[ngram][next_word] += 1
    return ngrams

def ngram_autocomplete(prefix, ngrams, n=5):
    words = prefix.split()
    ngram = tuple(words[-(n - 1):]) if len(words) >= n - 1 else tuple(words)
    suggestions = ngrams[ngram]
    return [word for word, _ in suggestions.most_common(n)]
```

```python
# Custom Autocorrect based on Edit Distance
def edit_distance_autocorrect(word, vocabulary):
    closest_word = min(vocabulary, key=lambda v: edit_distance(word, v))
    return closest_word
```
